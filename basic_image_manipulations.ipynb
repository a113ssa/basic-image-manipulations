```python
# Use the follwing IPython kernel option:
# InlineBackend.figure_formats: Set
# Default: set([‘png’])
# A set of figure formats to enable: ‘png’, ‘retina’, ‘jpeg’, ‘svg’, ‘pdf’

%config InlineBackend.figure_formats = 'retina'
```

### Create a simple greyscale image:

```python
img_gray = np.array([[0, 0, 255], [0, 255, 0], [255, 0, 0]], dtype = np.uint8)

# remove the axes
plt.axis('off')

# show image
# different colormaps: https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.imshow(img_gray, cmap='gray')
```

### Create a color image:

```python
img_color = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
                      [[255, 255, 0], [255, 0, 255], [0, 255, 255]],
                      [[255, 255, 255], [128, 128, 128], [0, 0, 0]]],
                      dtype = np.uint8)

# remove the axes
plt.axis('off')

# show image
plt.imshow(img_color)
```

### Read image from file:

```python
jpg = plt.imread('images/image1.jpg')
plt.axis('off')
plt.imshow(jpg)
```

### JPEG image shape:

```python
# 3 channels
jpg.shape
```

### JPEG image data type:

```python
print(jpg[0,0])

type(jpg[0,0,0])
```

### Cut image:

```python
cut = jpg[20:500, 50:500, :]
plt.axis('off')
plt.imshow(cut)
```

### Save image to file:

```python
plt.imsave('images/image1_cut.jpg', cut)
```
