```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load support methods for later use such as
# draw_subplots
# convert_png_to_jpg,
# print_6_first_and_last_rows_center_pixels
# print_pixels_comparison_description
# plot_histogram
# threshold_image
from support_methods import *

# Use the follwing IPython kernel option:
# InlineBackend.figure_formats: Set
# Default: set([‘png’])
# A set of figure formats to enable: ‘png’, ‘retina’, ‘jpeg’, ‘svg’, ‘pdf’
%config InlineBackend.figure_formats = 'retina'

jpg = plt.imread('images/image1.jpg')
```

### Apply average convolution filter:

```python
kernel_by_7 = np.ones((7,7), np.float32) / 49
blurred_by_7 = cv2.filter2D(jpg, ddepth=-1, kernel=kernel_by_7)

kernel_by_15 = np.ones((15,15), np.float32) / 225
blurred_by_15 = cv2.filter2D(jpg, ddepth=-1, kernel=kernel_by_15)

subplot_original = Subplot(cut_image_left_upper_quater(jpg), None, 'original')
subplot_blurred_by_7 = Subplot(cut_image_left_upper_quater(blurred_by_7), None, 'average blurred by 7')
subplot_blurred_by_15 = Subplot(cut_image_left_upper_quater(blurred_by_15), None, 'average blurred by 15')

draw_subplots([subplot_original, subplot_blurred_by_7, subplot_blurred_by_15], 1, 3, (10, 5))
print_6_first_and_last_rows_center_pixels([jpg, blurred_by_7, blurred_by_15],
                                          ['original', 'average blurred by 7', 'average blurred by 15'])
```

### Apply meadin convolution filter:

```python
# median blur works with original pixels
median_blurred_by_7 = cv2.medianBlur(jpg, 7)
median_blurred_by_15 = cv2.medianBlur(jpg, 15)

subplot_original = Subplot(cut_image_left_upper_quater(jpg), None, 'original')
subplot_median_blurred_by_7 = Subplot(cut_image_left_upper_quater(median_blurred_by_7), None, 'median blurred by 7')
subplot_median_blurred_by_15 = Subplot(cut_image_left_upper_quater(median_blurred_by_15), None, 'median blurred by 15')

draw_subplots([subplot_original, subplot_median_blurred_by_7, subplot_median_blurred_by_15], 1, 3, (10, 5))
print_6_first_and_last_rows_center_pixels([jpg, median_blurred_by_7, median_blurred_by_15],
                                          ['original', 'median blurred by 7', 'median blurred by 15'])
```

```python
gaussian_blurred_by_7 = cv2.GaussianBlur(jpg, (7, 7), cv2.BORDER_DEFAULT)
gaussian_blurred_by_15 = cv2.GaussianBlur(jpg, (15, 15), cv2.BORDER_DEFAULT)

subplot_original = Subplot(jpg), None, 'original')
subplot_gaussian_blurred_by_7 = Subplot(cut_image_left_upper_quater(gaussian_blurred_by_7), None, 'gaussian blurred by 7')
subplot_gaussian_blurred_by_15 = Subplot(cut_image_left_upper_quater(gaussian_blurred_by_15), None, 'gaussian blurred by 15')

draw_subplots([subplot_original, subplot_gaussian_blurred_by_7, subplot_gaussian_blurred_by_15], 1, 3, (10, 5))
print_6_first_and_last_rows_center_pixels([jpg, gaussian_blurred_by_7, gaussian_blurred_by_15],
                                          ['original', 'gaussian blurred by 7', 'gaussian blurred by 15'])
```

### Different blur filters comparison:

```python
draw_subplots([subplot_original, subplot_blurred_by_15, subplot_median_blurred_by_15, subplot_gaussian_blurred_by_15], 1, 4, (10, 5))
print_6_first_and_last_rows_center_pixels([jpg, blurred_by_15, median_blurred_by_15, gaussian_blurred_by_15],
                                          ['original', 'average blurred by 15', 'median blurred by 15', 'gaussian blurred by 15'])
```

### Perform rescaling

```python
downscaled = cv2.resize(jpg, None, fx=0.09, fy=0.09, interpolation=cv2.INTER_CUBIC)
upscaled = cv2.resize(jpg, None, fx=5, fy=5, interpolation=cv2.INTER_CUBIC)

subplot_original = Subplot(cut_image_left_upper_quater(jpg), None, 'original')
subplot_downscaled = Subplot(cut_image_left_upper_quater(downscaled), None, 'downscaled')
subplot_upscaled = Subplot(cut_image_left_upper_quater(upscaled), None, 'upscaled')

draw_subplots([subplot_original, subplot_downscaled, subplot_upscaled], 1, 3, (10, 5))
```
